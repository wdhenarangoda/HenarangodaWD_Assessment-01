```python
import hashlib
import time

# Hashing function
def calculate_block_hash(index, previous_hash, timestamp, data, proof):
    block_contents = f"{index}{previous_hash}{timestamp}{data}{proof}"
    return hashlib.sha256(block_contents.encode()).hexdigest()

# Create first block (Genesis)
index1 = 0
previous_hash1 = "0"
timestamp1 = time.time()
data1 = "Genesis Block"
proof1 = 0
hash1 = calculate_block_hash(index1, previous_hash1, timestamp1, data1, proof1)

print(f"Genesis Block Hash: {hash1}")

# Create second block (linked to Genesis)
index2 = 1
timestamp2 = time.time()
data2 = input("Enter transaction data for Block 2: ")
proof2 = 0
previous_hash2 = hash1  # Linking this block to the previous one
hash2 = calculate_block_hash(index2, previous_hash2, timestamp2, data2, proof2)

print(f"Block 2 Hash: {hash2}")
print(f"Block 2 Previous Hash: {previous_hash2}")
```

```python
import time

# Block class without automatic hash generation
class Block:
    def __init__(self, index, previous_hash, timestamp, data, proof):
        self.index = index
        self.previous_hash = previous_hash
        self.timestamp = timestamp
        self.data = data
        self.proof = proof
        self.hash = self.calculate_hash()  # Automatically calculate and store hash

    def calculate_hash(self):
        block_contents = f"{self.index}{self.previous_hash}{self.timestamp}{self.data}{self.proof}"
        return hashlib.sha256(block_contents.encode()).hexdigest()

# Blockchain class
class Blockchain:
    def __init__(self):
        self.chain = []
        self.create_genesis_block()

    def create_genesis_block(self):
        genesis_block = Block(0, "0", time.time(), "Genesis Block", 0)
        self.chain.append(genesis_block)

    def get_latest_block(self):
        return self.chain[-1]

    def add_block(self, data):
        latest_block = self.get_latest_block()
        new_index = latest_block.index + 1
        new_timestamp = time.time()
        new_proof = 0  # Proof of work to be added later
        new_block = Block(new_index, latest_block.hash, new_timestamp, data, new_proof)
        self.chain.append(new_block)


# MAIN EXECUTION

blockchain = Blockchain()

# Prompt user for input to add one block
user_input = input("Enter data to add to the blockchain: ")
blockchain.add_block(user_input)

# Display the blockchain
print("\nCurrent Blockchain:")
for block in blockchain.chain:
    print(f"Block-{block.index}")
    print(f"  Data        : {block.data}")
    print(f"  Hash        : {block.hash}")
    print(f"  Prev Hash   : {block.previous_hash}\n")
```
