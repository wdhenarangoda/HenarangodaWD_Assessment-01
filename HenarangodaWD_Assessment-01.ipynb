```python
import time

# Create a unique identifier using timestamp
def generate_unique_block_id(index, timestamp):
    return f"Block-{index}-{int(timestamp)}"

# Example usage
index = 0
timestamp = time.time()
unique_id = generate_unique_block_id(index, timestamp)
print("Unique Block ID:", unique_id)
```
